```python
import pandas as pd
```

```python
df = pd.read_csv('쓰레기수거현황.csv',engine='python', encoding='utf-8', sep=",")
```

# 데이터 전처리

```python
df
```

```python
df.reset_index(inplace = True)
```

```python
df = df.rename(columns=df.iloc[0])
```

```python
df = df.drop(0, axis = 0)
```

```python
df
```

```python
# 분석에 사용할 컬럼만 분리
df2 = df.iloc[:, [0, 2, 3, 4, 8]]
```

```python
df2
```

```python
# 컬럼 생성을 위해 컬럼별 type 변경
df2 = df2.astype({'인구 (명)':'int','면적 (㎢)':'float', '소계':'float'})
df2.info()
```

```python
# 인구별 쓰레기 배출량 컬럼 생성
df2['인구별'] = df2['소계']/df2['인구 (명)']
```

```python
# 면적별 쓰레기 배출량 컬럼 생성
df2['면적별'] = df2['소계']/df2['면적 (㎢)']
```

```python
df2 = df2.set_index('시점')
```

```python
# 년도별 데이터 분리
df2010 = df2[df2.index == '2010']
df2011 = df2[df2.index == '2011']
df2012 = df2[df2.index == '2012']
df2013 = df2[df2.index == '2013']
df2014 = df2[df2.index == '2014']
df2015 = df2[df2.index == '2015']
df2016 = df2[df2.index == '2016']
df2017 = df2[df2.index == '2017']
df2018 = df2[df2.index == '2018']
df2019 = df2[df2.index == '2019']
df2020 = df2[df2.index == '2020']
```

```python
# 총쓰레기배출량 데이터만 추출
df2010 = df2010.set_index('자치구별(2)')['소계']
df2011 = df2011.set_index('자치구별(2)')['소계']
df2012 = df2012.set_index('자치구별(2)')['소계']
df2013 = df2013.set_index('자치구별(2)')['소계']
df2014= df2014.set_index('자치구별(2)')['소계']
df2015 = df2015.set_index('자치구별(2)')['소계']
df2016 = df2016.set_index('자치구별(2)')['소계']
df2017= df2017.set_index('자치구별(2)')['소계']
df2018 = df2018.set_index('자치구별(2)')['소계']
df2019 = df2019.set_index('자치구별(2)')['소계']
df2020 = df2020.set_index('자치구별(2)')['소계']
```

```python
# 년도별 총 쓰레기배출량 데이터 병합
year_df = pd.concat([df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020],axis=1)
```

```python
# 컬럼명 변경
year_df.columns = [x for x in range(2010,2021)]
year_df
```
